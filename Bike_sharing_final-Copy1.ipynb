# Above categorial variabe we can transform to dummy variable to further conclude the analysis.

```python
df['season']=df['season'].astype('category')
df['weathersit']=df['weathersit'].astype('category')
df['mnth']=df['mnth'].astype('category')
df['weekday']=df['weekday'].astype('category')
df.info() # to check the Dtype
```

### Now we need to convert to columns for the above category variable and drop the orignal columns

```python
df=pd.get_dummies(data=df,drop_first=True)
print(df.info())
df.head()
```

### Splitting the train nd test date 70:30

```python
from sklearn.model_selection import train_test_split

np.random.seed(0)
df_train,df_test=train_test_split(df,train_size=0.7,test_size=0.3,random_state=1000)
```

```python
### Check the data size

print(df_train.shape)
print(df_test.shape)
```

## Rescaling the int data 

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
v_num = ['atemp', 'hum','temp','windspeed','cnt']
df_train[v_num]=scaler.fit_transform(df_train[v_num])
```

#### Lets build the model X and Y train and Build the Stats model

```python
y_train = df_train.pop('cnt')
X_train = df_train
```

```python
## Import Libraries 
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
print(X_train.shape)
y_train.shape
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
rfe = RFE(estimator= lm, n_features_to_select=15)
```

```python
rfe=rfe.fit(X_train,y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

```python

X_train_rfe = X_train[col]
X_train_rfe.columns
```

### Build the Linear Model and cmpare the Leniar regression Model with the value of viF model and try to eleminate the unwanted columns by elemination method and create various models.

# Build First  Model 

```python
import statsmodels.api as sm  

X_train_lm1 = sm.add_constant(X_train_rfe)
```

```python
lr1=sm.OLS(y_train,X_train_lm1).fit()  ### Run the Model
```

```python
print(lr1.summary())
```

#### Build the VIF Model

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif=pd.DataFrame()
vif['Features']=X_train_rfe.columns
vif['VIF']=[variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

  ###  We can see the temp and hum have very high VIF hence good idea to drop but temp have directly related to season which we concluded above in EDA hence first removing HUM and hum have -ve coef correlation so removing this columns first .

# Builld Second Model

```python
X_train_rfe2 = X_train_rfe.drop(["hum"], axis = 1)
```

```python
X_train_lm2 = sm.add_constant(X_train_rfe2)

lr2=sm.OLS(y_train,X_train_lm2).fit()

print(lr2.summary())
```

```python
# Create a dataframe that will contain the names of all the feature variables and their respective VIFs
vif=pd.DataFrame()
vif['Features']=X_train_rfe2.columns
vif['VIF']=[variance_inflation_factor(X_train_rfe2.values, i) for i in range(X_train_rfe2.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

#### Lets drop Season 3 having high VIF



# Build Third Model

```python
X_train_rfe3=X_train_rfe2.drop(['season_3'],axis=1)
```

```python
X_train_lm3=sm.add_constant(X_train_rfe3)

lr3=sm.OLS(y_train,X_train_lm3).fit()


print(lr3.summary())
```

```python
vif=pd.DataFrame()
vif['Features']=X_train_rfe3.columns
vif['VIF']=[variance_inflation_factor(X_train_rfe3.values,i) for i in range(X_train_rfe3.shape[1])]
vif['VIF']=round(vif["VIF"],2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

#### Now lets drop Both Windspeed and Season 2 having high VIF number

# Build fourth Model

```python
X_train_rfe4=X_train_rfe3.drop(['windspeed','season_2'],axis=1)
```

```python
X_train_lm4=sm.add_constant(X_train_rfe4)

lr4=sm.OLS(y_train,X_train_lm4).fit()

print(lr4.summary())
```

```python
vif=pd.DataFrame()
vif['Features']=X_train_rfe4.columns
vif['VIF']=[variance_inflation_factor(X_train_rfe4.values,i) for i in range(X_train_rfe4.shape[1])]
vif['VIF']=round(vif['VIF'],2)
vif
```

### Month 8 and 10 is having high P value we can delete this two columns

# Build Model Five

```python
X_train_rfe5=X_train_rfe4.drop(['mnth_8','mnth_10'],axis=1)
```

```python
X_train_lm5=sm.add_constant(X_train_rfe5)

lr5=sm.OLS(y_train,X_train_lm5).fit()

print(lr5.summary())
```

```python
vif=pd.DataFrame()
vif['Features']=X_train_rfe5.columns
vif['VIF']=[variance_inflation_factor(X_train_rfe5.values,i) for i in range(X_train_rfe5.shape[1])]
vif['VIF']=round(vif['VIF'],2)
vif
```

## This above model looks good to me as I can now conclude the both R-squared and adjusted r value are very near(.799 and .795) and also P value is 0 for almost all and as studed in theory sessions its good to consider any value at.05 hence I will go ahead with above model.

# Residual Analysis

```python
y_train_cnt = lr5.predict(X_train_lm5)
```

```python
res = y_train-y_train_cnt

fig = plt.figure()
sns.distplot((res), bins = 20)
fig.suptitle('Error Terms', fontsize = 19)                  # Plot heading 
plt.xlabel('Errors', fontsize = 19)     
```

### From the above plat we can say that our analysis is correct its linear plotted and peak at 0.0 hence our analysis is correct.
